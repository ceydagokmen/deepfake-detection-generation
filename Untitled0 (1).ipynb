```python
import torch
import torchvision.transforms as transforms
from torchvision.datasets import ImageFolder
from torch.utils.data import DataLoader, Subset
import numpy as np

# Görüntü ön işleme
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406],
                         [0.229, 0.224, 0.225])
])

# Dataset yükle
train_data = ImageFolder("/content/deepfake_project/data/real_vs_fake/real-vs-fake/train", transform=transform)
valid_data = ImageFolder("/content/deepfake_project/data/real_vs_fake/real-vs-fake/valid", transform=transform)

# Her sınıftan 5000 görüntü al (toplam 10000)
def get_subset(dataset, per_class=5000):
    indices = []
    for class_idx in range(len(dataset.classes)):
        class_indices = [i for i, (_, label) in enumerate(dataset.samples) if label == class_idx]
        indices += class_indices[:per_class]
    return Subset(dataset, indices)

train_subset = get_subset(train_data, per_class=5000)
valid_subset = get_subset(valid_data, per_class=2000)

# DataLoader
train_loader = DataLoader(train_subset, batch_size=32, shuffle=True, num_workers=2)
valid_loader = DataLoader(valid_subset, batch_size=32, shuffle=False, num_workers=2)

print("Sınıflar:", train_data.classes)
print("Eğitim seti:", len(train_subset), "görüntü")
print("Validasyon seti:", len(valid_subset), "görüntü")
```

```python
import torch.nn as nn
from torchvision import models

# EfficientNet'i yükle (ImageNet'te önceden eğitilmiş)
model = models.efficientnet_b0(weights="IMAGENET1K_V1")

# Son katmanı 2 sınıf için değiştir (fake / real)
model.classifier[1] = nn.Linear(model.classifier[1].in_features, 2)

# GPU'ya taşı
device = torch.device("cuda")
model = model.to(device)

print("Model hazır!")
print(f"Toplam parametre: {sum(p.numel() for p in model.parameters()):,}")
```

```python
import torch.optim as optim
from torch.optim.lr_scheduler import StepLR

# Kayıp fonksiyonu ve optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.0001)
scheduler = StepLR(optimizer, step_size=3, gamma=0.5)

# Eğitim fonksiyonu
def train_epoch(model, loader, optimizer, criterion):
    model.train()
    total_loss, correct, total = 0, 0, 0
    for images, labels in loader:
        images, labels = images.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
        correct += (outputs.argmax(1) == labels).sum().item()
        total += labels.size(0)
    return total_loss / len(loader), correct / total

# Validasyon fonksiyonu
def validate(model, loader, criterion):
    model.eval()
    total_loss, correct, total = 0, 0, 0
    with torch.no_grad():
        for images, labels in loader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            loss = criterion(outputs, labels)
            total_loss += loss.item()
            correct += (outputs.argmax(1) == labels).sum().item()
            total += labels.size(0)
    return total_loss / len(loader), correct / total

print("Eğitim fonksiyonları hazır, başlatılıyor...")

# Eğitim döngüsü (5 epoch)
history = {"train_loss": [], "train_acc": [], "val_loss": [], "val_acc": []}

for epoch in range(5):
    train_loss, train_acc = train_epoch(model, train_loader, optimizer, criterion)
    val_loss, val_acc = validate(model, valid_loader, criterion)
    scheduler.step()

    history["train_loss"].append(train_loss)
    history["train_acc"].append(train_acc)
    history["val_loss"].append(val_loss)
    history["val_acc"].append(val_acc)

    print(f"Epoch {epoch+1}/5 | Train Loss: {train_loss:.4f} | Train Acc: {train_acc*100:.2f}% | Val Loss: {val_loss:.4f} | Val Acc: {val_acc*100:.2f}%")

print("\nEğitim tamamlandı!")
```

```python
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns

model.eval()
all_preds, all_labels = [], []

test_data = ImageFolder(
    "/content/deepfake_project/data/real_vs_fake/real-vs-fake/test",
    transform=transform
)
test_subset = get_subset(test_data, per_class=1000)
test_loader = DataLoader(test_subset, batch_size=32, shuffle=False, num_workers=2)

with torch.no_grad():
    for images, labels in test_loader:
        images = images.to(device)
        outputs = model(images)
        preds = outputs.argmax(1).cpu().numpy()
        all_preds.extend(preds)
        all_labels.extend(labels.numpy())

# Confusion matrix
cm = confusion_matrix(all_labels, all_preds)
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=["Fake", "Real"],
            yticklabels=["Fake", "Real"])
plt.title("Confusion Matrix")
plt.ylabel("Gerçek Etiket")
plt.xlabel("Tahmin")
plt.savefig("/content/deepfake_project/results/confusion_matrix.png", dpi=150)
plt.show()

# Detaylı rapor
print("\nSınıflandırma Raporu:")
print(classification_report(all_labels, all_preds, target_names=["Fake", "Real"]))
```

```python
def face_swap(source_img, target_img):
    # PIL'den numpy'a çevir
    src = cv2.cvtColor(np.array(source_img), cv2.COLOR_RGB2BGR)
    tgt = cv2.cvtColor(np.array(target_img), cv2.COLOR_RGB2BGR)

    # Yüzleri tespit et
    src_faces = app.get(src)
    tgt_faces = app.get(tgt)

    if len(src_faces) == 0:
        return None, "❌ Kaynak fotoğrafta yüz bulunamadı!"
    if len(tgt_faces) == 0:
        return None, "❌ Hedef fotoğrafta yüz bulunamadı!"

    # Yüzü değiştir
    result = tgt.copy()
    result = swapper.get(result, tgt_faces[0], src_faces[0], paste_back=True)

    # BGR'dan RGB'ye çevir
    result_rgb = cv2.cvtColor(result, cv2.COLOR_BGR2RGB)
    result_pil = Image.fromarray(result_rgb)

    return result_pil, "✅ Yüz değiştirme başarılı!"

print("Face swap fonksiyonu hazır!")
```

```python
def detect(image):
    img_tensor = transform_single(image).unsqueeze(0).to(device)
    with torch.no_grad():
        output = model(img_tensor)
        probs = torch.softmax(output, dim=1)[0]
    fake_prob = probs[0].item()
    real_prob = probs[1].item()
    if fake_prob > real_prob:
        label = "🔴 SAHTE (Deepfake)"
        confidence = fake_prob
    else:
        label = "🟢 GERÇEK"
        confidence = real_prob
    return {f"{label} — Güven: {confidence*100:.1f}%": confidence, "Karşı ihtimal": 1 - confidence}

def generate(_):
    seed = np.random.randint(0, 99999)
    z = torch.from_numpy(np.random.RandomState(seed).randn(1, G.z_dim)).cuda().float()
    c = torch.zeros([1, G.c_dim]).cuda()
    with torch.no_grad():
        img = G(z, c, truncation_psi=0.7)
    img = (img.permute(0, 2, 3, 1) * 127.5 + 128).clamp(0, 255).to(torch.uint8)
    return Image.fromarray(img[0].cpu().numpy())

def swap_and_detect(source_img, target_img):
    swapped, msg = face_swap(source_img, target_img)
    if swapped is None:
        return None, msg, {}
    result = detect(swapped)
    return swapped, msg, result

with gr.Blocks(title="Deepfake Sistemi") as demo:
    gr.Markdown("# 🎭 Deepfake Algılama ve Üretme Sistemi")

    with gr.Tab("🔍 Deepfake Algılama"):
        gr.Markdown("Bir yüz fotoğrafı yükle — gerçek mi sahte mi olduğunu söyleyelim.")
        with gr.Row():
            inp = gr.Image(type="pil", label="Fotoğraf Yükle")
            out = gr.Label(label="Sonuç")
        btn1 = gr.Button("Analiz Et", variant="primary")
        btn1.click(fn=detect, inputs=inp, outputs=out)

    with gr.Tab("🎨 Sahte Yüz Üret (StyleGAN2)"):
        gr.Markdown("Butona bas — hiç var olmamış bir yüz üretilsin.")
        out_img = gr.Image(label="Üretilen Sahte Yüz")
        btn2 = gr.Button("Yeni Yüz Üret 🎲", variant="primary")
        btn2.click(fn=generate, inputs=btn2, outputs=out_img)

    with gr.Tab("🔄 Yüz Değiştir + Tespit"):
        gr.Markdown("İki fotoğraf yükle — birinin yüzü diğerine geçsin, sonra tespit edilsin.")
        with gr.Row():
            src_inp = gr.Image(type="pil", label="Kaynak Yüz (kimin yüzü alınacak)")
            tgt_inp = gr.Image(type="pil", label="Hedef (yüz kimin üzerine gidecek)")
        btn3 = gr.Button("Yüz Değiştir ve Tespit Et", variant="primary")
        with gr.Row():
            swap_out = gr.Image(label="Sonuç Görüntü")
            swap_msg = gr.Textbox(label="Durum")
            detect_out = gr.Label(label="Tespit Sonucu")
        btn3.click(fn=swap_and_detect, inputs=[src_inp, tgt_inp], outputs=[swap_out, swap_msg, detect_out])

demo.launch(share=True)
```

```python
app_code = '''
import gradio as gr
import torch
import torchvision.transforms as transforms
from torchvision import models
import torch.nn as nn
from PIL import Image
import numpy as np
import sys
sys.path.insert(0, './stylegan2-ada-pytorch')
import pickle

# Cihaz
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Algılama modeli
model = models.efficientnet_b0(weights=None)
model.classifier[1] = nn.Linear(model.classifier[1].in_features, 2)
model.load_state_dict(torch.load("deepfake_detector_v2.pth", map_location=device))
model = model.to(device)
model.eval()

transform_single = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])

# StyleGAN2
with open("ffhq.pkl", "rb") as f:
    G = pickle.load(f)["G_ema"].to(device)

def detect(image):
    img_tensor = transform_single(image).unsqueeze(0).to(device)
    with torch.no_grad():
        output = model(img_tensor)
        probs = torch.softmax(output, dim=1)[0]
    fake_prob = probs[0].item()
    real_prob = probs[1].item()
    if fake_prob > real_prob:
        label = "🔴 SAHTE (Deepfake)"
        confidence = fake_prob
    else:
        label = "🟢 GERÇEK"
        confidence = real_prob
    return {f"{label} — Güven: {confidence*100:.1f}%": confidence, "Karşı ihtimal": 1 - confidence}

def generate(_):
    seed = np.random.randint(0, 99999)
    z = torch.from_numpy(np.random.RandomState(seed).randn(1, G.z_dim)).to(device).float()
    c = torch.zeros([1, G.c_dim]).to(device)
    with torch.no_grad():
        img = G(z, c, truncation_psi=0.7)
    img = (img.permute(0, 2, 3, 1) * 127.5 + 128).clamp(0, 255).to(torch.uint8)
    return Image.fromarray(img[0].cpu().numpy())

with gr.Blocks(title="Deepfake Sistemi") as demo:
    gr.Markdown("# 🎭 Deepfake Algılama ve Üretme Sistemi")
    with gr.Tab("🔍 Deepfake Algılama"):
        gr.Markdown("Bir yüz fotoğrafı yükle — gerçek mi sahte mi olduğunu söyleyelim.")
        with gr.Row():
            inp = gr.Image(type="pil", label="Fotoğraf Yükle")
            out = gr.Label(label="Sonuç")
        btn1 = gr.Button("Analiz Et", variant="primary")
        btn1.click(fn=detect, inputs=inp, outputs=out)
    with gr.Tab("🎨 Sahte Yüz Üret (StyleGAN2)"):
        gr.Markdown("Butona bas — hiç var olmamış bir yüz üretilsin.")
        out_img = gr.Image(label="Üretilen Sahte Yüz")
        btn2 = gr.Button("Yeni Yüz Üret 🎲", variant="primary")
        btn2.click(fn=generate, inputs=btn2, outputs=out_img)

demo.launch()
'''

with open("/content/repo/app.py", "w") as f:
    f.write(app_code)

print("app.py oluşturuldu!")
```
